# Background information on this data
The stimulus is a sequence of random fake data that was generated using a Gaussian distribution. This is commonly called "Gaussian White Noise", or noise that is Gaussianly distributed. 

The cell's response was created by convolving a filter of size (6x1) with the stimulus. To get a visual intuition as to what convolution is, please see this [page on convolution](https://en.wikipedia.org/wiki/Convolution).

$\text{filter} = \frac{1}{n_{sp}}\sum_{i=1}^{T}y_i\bar{x_i}$
